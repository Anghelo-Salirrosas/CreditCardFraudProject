## 3. Data Cleaning
We're going to discard columns with a lot of missing values

```python
## First we discard irrelevants columns with more than 30% missing values, tha majority are in this range
discards = original_data.columns[44:90].to_list()
discards.append('OWN_CAR_AGE')

## Second we discard columns with categorial data with too many categories 
discards.append('ORGANIZATION_TYPE')

## Third we discard identifier columns
discards.append('SK_ID_CURR') 

```

Since "OCCUPATION_TYPE" have 18 categories, we are going to group less commons and missing values in a category "otros"

```python
selected_features['OCCUPATION_TYPE'].isnull().sum()
```

```python
selected_features['OCCUPATION_TYPE'].value_counts()
```

```python
col = "OCCUPATION_TYPE"
umbral = 6000

# Crear copia para evitar el warning
selected_features = selected_features.copy()

# CategorÃ­as principales
categorias_principales = (
    selected_features[col].value_counts()[selected_features[col].value_counts() >= umbral].index
)

# Agrupar categorÃ­as raras y NaN
selected_features.loc[:, col] = selected_features[col].where(
    selected_features[col].isin(categorias_principales), "Otros"
)
selected_features.loc[:, col] = selected_features[col].fillna("Otros")
```

```python
selected_features['OCCUPATION_TYPE'].value_counts()
```

```python
cat_columns_no_ocupation = selected_features.select_dtypes(include=['object']).columns.tolist()
cat_columns_no_ocupation.remove('OCCUPATION_TYPE')
cat_columns_no_ocupation
```

```python
imputer = SimpleImputer(strategy='most_frequent')
selected_features.loc[:, cat_columns_no_ocupation] = imputer.fit_transform(selected_features[cat_columns_no_ocupation])
```

```python
selected_features['OCCUPATION_TYPE'].isnull().sum()
```

```python
selected_features.isnull().sum().sum()
```

#### Dropping incompatible data in certain columns
Probably in the 'DAYS_EMPLOYED' column, for unemployed people was used the value 365243. 

Since we don't want to lose this data, we are going to separate that data to another column called 'IS_EMPLOYED'. 

And we will changes the values 365243 to 0 in the original column.

```python
selected_features['IS_UNEMPLOYED'] = selected_features['DAYS_EMPLOYED'].apply(lambda x: 0 if x == 365243 else 1)
```

```python
selected_features['DAYS_EMPLOYED'] = selected_features['DAYS_EMPLOYED'].replace(365243, 0)
```

```python
# To make an analysis of the data
selected_features.to_csv('../application_data/cleaned_data_before_transform.csv', index=False)
```

## 5. Transforming data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
transformed_data = selected_features.copy()
```

```python
categorical_data = transformed_data.select_dtypes(include=['object']).columns
categorical_data
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore",sparse_output=False), categorical_data)  # OneHot en categÃ³ricas
    ],
    remainder="passthrough"  # Passthrough para numÃ©ricas
)

X_transformed = preprocessor.fit_transform(transformed_data)
feature_names = preprocessor.get_feature_names_out()
```

```python
no_scaled_data = pd.DataFrame(X_transformed, columns=feature_names, index=selected_features.index)
# Quitar prefijo "cat__" y "remainder__"
clean_names = [name.split("__")[-1] for name in no_scaled_data.columns]
no_scaled_data.columns = clean_names
```

```python
no_scaled_data.to_csv('../application_data/cleaned_data_no_scaled.csv', index=False)
```

```python
num_cols = [col for col in no_scaled_data.select_dtypes(include=np.number).columns if col != "TARGET"]
scaled_data = no_scaled_data.copy()
# Inicializar StandardScaler
scaler = StandardScaler()

# Ajustar y transformar
scaled_data[num_cols] = scaler.fit_transform(no_scaled_data[num_cols])
```

```python
scaled_data.to_csv('../application_data/cleaned_data_scaled.csv', index=False)
```
