1. Distribucin de clientes por gnero

```python
%%sql 
SELECT 
    CODE_GENDER, 
    COUNT(*) AS cantidad, 
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM credit_card_db), 2) AS porcentaje
FROM credit_card_db
GROUP BY CODE_GENDER;
```

2. Distribucin de cliente por genero en caso de fraude

```python
%%sql 
SELECT 
    CODE_GENDER, 
    COUNT(*) AS cantidad, 
    ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM credit_card_db WHERE TARGET == 1), 2) AS porcentaje
FROM credit_card_db
WHERE TARGET == 1
GROUP BY CODE_GENDER;
```

3. Distribucin de fraude por tipo de contrato

```python
%%sql
SELECT 
    NAME_CONTRACT_TYPE, 
    COUNT(*) AS cantidad
FROM credit_card_db
WHERE TARGET == 1
GROUP BY NAME_CONTRACT_TYPE;
```

4. Distribucin de fraude por educacin

```python
%%sql
SELECT
    NAME_EDUCATION_TYPE,
    COUNT(*) AS cantidad
FROM credit_card_db
WHERE TARGET == 1
GROUP BY NAME_EDUCATION_TYPE;
```

5. Distribucin de fraude por edad

```python
%%sql
SELECT
    ROUND(DAYS_BIRTH / -365.0, 0) AS edad,
    COUNT(*) AS cantidad
FROM credit_card_db
WHERE TARGET == 1
GROUP BY EDAD
ORDER BY cantidad DESC
LIMIT 5;
```

6. Distribucin de fraude por cantidad de meses que se cambi el telfono

```python
%%sql
SELECT
    ROUND(DAYS_LAST_PHONE_CHANGE / 30,0) AS meses_ultimo_cambio,
    COUNT(*) AS cantidad
FROM credit_card_db
WHERE TARGET == 1
GROUP BY meses_ultimo_cambio
ORDER BY cantidad DESC
LIMIT 10;
```

7. Distribucin de fraude por tipo trabajo

```python
%%sql
SELECT 
    OCCUPATION_TYPE,
    COUNT(*) AS cantidad
FROM credit_card_db
WHERE TARGET == 1
GROUP BY OCCUPATION_TYPE
ORDER BY cantidad DESC
LIMIT 10;
```

8. Distribucion de fraude por da de la semana

```python
%%sql
SELECT 
    WEEKDAY_APPR_PROCESS_START,
    COUNT(*) AS cantidad
FROM credit_card_db
WHERE TARGET == 1
GROUP BY WEEKDAY_APPR_PROCESS_START
ORDER BY cantidad DESC
LIMIT 10;
```
